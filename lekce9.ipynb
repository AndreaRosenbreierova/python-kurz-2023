```python
import pandas as pd
staty = pandas.read_json("staty.json")
staty.set_index("name", inplace=True)
#staty.loc["Czech Republic"]

```

```python
pd.set_option('display.max_rows', 500)
pidistaty = staty[staty["population"] < 1000]

```

```python
pd.set_option('display.max_rows', 500)
velke_staty = staty[staty["area"] > 500000]
velke_staty
```

```python
lidnate_evropske_staty = staty[(staty["population"] > 20_000_000) & (staty["region"] == "Europe")]
lidnate_evropske_staty
```

```python
vybrane_staty = staty[(staty["population"] > 20_000_000) | (staty["region"] == "Europe")]
vybrane_staty
```

```python
zapadni_vychodni_evropa = staty[staty["subregion"].isin(["Western Europe", "Eastern Europe"])]
zapadni_vychodni_evropa
```
